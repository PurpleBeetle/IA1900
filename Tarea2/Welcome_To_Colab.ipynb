```
import pandas as pd
import numpy as np


# Ejercicio 1: An√°lisis de Ventas
# Utilice el archivo ventas.csv para responder lo siguiente:
#1. Cargue los datos en un DataFrame
ventas = pd.read_csv('ventas.csv')
ventas
```

```
#2. Calcule la cantidad total de productos vendidos por categor√≠a
ventas_por_categoria = ventas.groupby('Producto')['Cantidad'].sum()
ventas_por_categoria

# aqui sumamos la grupacion de productos por cantidad
```

```
#3. Determine cu√°l es el producto con el mayor total de ventas.

producto_top = ventas_por_categoria.idxmax()
producto_top

# Aqui usamos idxmax para el indice del df de ventas
```

```
#4. Encuentre el precio promedio de los productos vendidos.
precio_promedio = ventas.groupby('Producto')['Precio_Unitario'].mean()
precio_promedio

# se una mean() para conseguir el promedio de ventas por las agrupaciones de producto
```

```
# Ejercicio 2: An√°lisis del Clima
# Utilice el archivo clima.csv para analizar los datos clim√°ticos:
# 1.Cargue el dataset y convi√©rtelo en un DataFrame.

clima = pd.read_csv('clima.csv')
clima
```

```
# 2.Calcule la temperatura promedio de cada ciudad.
clima_promedio = clima.groupby('Ciudad')['Temperatura'].mean()
clima_promedio

# Usamos mean() para calcular el promedio de temperatura por la agrupacion de ciudades
```

```
# 3.Determine el registro con la temperatura m√°s alta y el m√°s bajo en el dataset
registro_max = clima.loc[clima['Temperatura'].idxmax()]
registro_min = clima.loc[clima['Temperatura'].idxmin()]
registro_max


#conseguimos el indice maximo y mino del row por Temperatura
```

```
registro_min
```

```
# 4. Identifique qu√© ciudad tuvo la temperatura m√°s alta y cu√°l la m√°s baja en todo el dataset.
ciudad_mas_caliente = registro_max['Ciudad']
ciudad_mas_fria = registro_min['Ciudad']
ciudad_mas_caliente

#Ya teniendo indices, solo se hacen la referencia con la ciudad.
```

```
ciudad_mas_fria
```

```
# 5.Encuentre cu√°ntos registros tienen una temperatura mayor a 30¬∞C.

temp_mayor_30 = clima[clima['Temperatura'] > 30].reset_index(drop=False)
temp_mayor_30

# filtramos todas las temperaturas mayor a 30, y hacemos reset el indice para reconocer la cantidad
```

```
# 6.Cuenta cu√°ntos d√≠as en total hay registrados por cada ciudad.

dias_por_ciudad = clima['Ciudad'].value_counts()
dias_por_ciudad

# usamos value_counts para contar los valores de Ciudad
```

```
# Ejercicio 3: Estad√≠sticas de Calificaciones
# Utilice el archivo calificaciones.csv para analizar el rendimiento de los estudiantes:

#1. Cargue el dataset y muestre las primeras filas.
calificaciones = pd.read_csv('calificaciones.csv')
calificaciones.head()

# head() nos ayuda imprimir los primeros 5 registros.
```

```
#2. Calcule el promedio de calificaciones por materia.
promedio_materia = calificaciones.groupby('Materia')['Calificaci√≥n'].mean()
promedio_materia

#similar a otros ejercicios, tenemos el promedio con mean() de las calificaciones de la agrupacion de materias
```

```
#3. Identifique el estudiante con el promedio m√°s alto.
promedio_lato = calificaciones.loc[calificaciones['Calificaci√≥n'].idxmax()]
estudiantes_promedio_alto = promedio_lato['Estudiante']
estudiantes_promedio_alto

#se filtra el nombre de estudiante por el indice del filtro de calificaciones
```

```
# 4. Agrupa las calificaciones por estudiante y calcule el promedio de cada uno.
estudiantes_promedio = calificaciones.groupby('Estudiante')['Calificaci√≥n'].mean()
estudiantes_promedio

# el promedio de calificaciones por la agrupacion de estudiantes
```

```
# 5. Identifique cu√°ntos estudiantes tienen un promedio superior a 85.
estudiantes_sobresalientes = estudiantes_promedio[estudiantes_promedio > 85].count()
estudiantes_sobresalientes

# se cuentan cuantos estudiantes con la condicion > 85 hay en en el df
```

```
# 6. Encuentre la materia con la mayor cantidad de calificaciones registradas.
materia_mas_calificaciones = calificaciones['Materia'].value_counts().idxmax()
materia_mas_calificaciones

#Se consigue el indice de la materia con el value_counts()
```
